```python
import pandas as pd
```

```python
df=pd.read_csv("Updated_employee_data.csv")
performance_df = pd.read_csv("performance_data.csv")
merged_df=pd.merge(df, performance_df, on='EmployeeID', how='left')
merged_df.to_csv()
```

```python
# prompt: merged_df.to_csv() convert it in csv

merged_df.to_csv('merged_employee_data.csv', index=False)
```

```python
# Filter rows based on multiple conditions involving columns like Salary, ExperienceYears, Location, and PerformanceRating.

merged_df[['Salary', 'ExperienceYears', 'Location', 'PerformanceRating']]
```

```python
# Filter employees with Salary > â‚¹60,000 and experience > 5 years.

merged_df[(merged_df['Salary']>60000)&(merged_df['ExperienceYears']>5)]
```

```python
# Select rows where Age < 30 and Department is 'HR'.

merged_df[(merged_df['Age']<30)&(merged_df['Department']=='HR')]
```

```python
# Get all rows where PerformanceRating is 'Good' or 'Excellent'.

merged_df[merged_df['PerformanceRating'].isin(['Good', 'Excellent'])]
```

```python
# Employees from Pune or Bangalore with Bonus > â‚¹15,000.

merged_df[(merged_df['Location'].isin(['Pune', 'Bangalore']))&(merged_df['Salary']>15000)]
```

```python
# AgeGroup is 'Senior' and LastPromotionYear < 2020.

merged_df[(merged_df['AgeGroup'] == 'Senior') & (merged_df['LastPromotionYear'] < 2020)]
```

```python
# Employees whose Salary is between â‚¹50,000 and â‚¹80,000.

merged_df[(merged_df['Salary']>50000)&(merged_df['Salary']<80000)]
```

```python
# Filter employees working in 'Data Science' with rating not 'Poor'.

merged_df[(merged_df['Department']=='Data Science')&(merged_df['PerformanceRating']!='Poor')]
```

```python
# Filter all employees aged under 35 from Mumbai.

merged_df[(merged_df['Age']<35)&(merged_df['Location']=='Mumbai')]
```

```python
# Employees with less than 3 years experience or Salary > â‚¹90,000.

merged_df[(merged_df['ExperienceYears']<3)&(merged_df['Salary']>90000)]
```

```python
# Employees who are from Delhi, are over 40 years old, and have a rating of "Good".

merged_df[(merged_df['Location']=='Delhi')&(merged_df['Age']>40)&(merged_df['PerformanceRating']=='Good')]
```

```python
# Employees with experience > 7 years and AgeGroup is "Senior".

merged_df[(merged_df['ExperienceYears']>7)&(merged_df['AgeGroup']=='Senior')]
```

```python
# prompt: # Employees with salary above average salary and bonus below 10k.
k=merged_df[['Salary']].mean().iloc[0]
# merged_df[(merged_df['Salary']>k)&(merged_df['Salary']<10000)]

merged_df[(merged_df['Salary']>k)&(merged_df['Salary']<10000)]
```

```python
# Employees from Marketing or Sales with Rating not in ['Poor', 'Average'].

merged_df[(merged_df['Department'].isin(['Marketing',  'Sales']))&(~merged_df['PerformanceRating'].isin(['Poor','Average']))]
```

```python
merged_df[~merged_df['Location'].isin(['Bangalore', 'Pune', 'Delhi'])]
```

```python
# prompt: Employees promoted in the last 2 years and earning less than 70k.
# from the dataset


merged_df[(merged_df['LastPromotionYear'] <= 2021) & (merged_df['Salary'] < 70000)]
```

```python
# Employees from 'HR' or 'Finance' departments with Age < 30.

merged_df[(merged_df['Department'].isin(['HR', 'Finance']))&(merged_df['Age']<30)]
```

```python
# prompt: Filter based on a list of names or departments using isin().

# Filter based on a list of names or departments using isin().

names_to_filter = ['Alice', 'Bob'] # Replace with actual names
departments_to_filter = ['IT', 'Marketing'] # Replace with actual departments

filtered_by_names = merged_df[merged_df['Name'].isin(names_to_filter)]
filtered_by_departments = merged_df[merged_df['Department'].isin(departments_to_filter)]

print("Filtered by Names:")
print(filtered_by_names)
print("\nFiltered by Departments:")
filtered_by_departments
```

```python
# Find employees who are not eligible for promotion but rated 'Excellent'.

merged_df[(merged_df['EligibleForPromotion']=='No')&(merged_df['PerformanceRating']=="Excellent")]
```

```python
# Employees who have a non-null LastPromotionYear and salary > 90k.
merged_df[(merged_df['LastPromotionYear'].isnull())&(merged_df['Salary']>90000)]
```

```python
# prompt: # Filter those with both high salary and experience in top 10% of dataset.

# Filter those with both high salary and experience in top 10% of dataset.

# Calculate the 90th percentile for Salary and ExperienceYears
salary_threshold = merged_df['Salary'].quantile(0.9)
experience_threshold = merged_df['ExperienceYears'].quantile(0.9)


# Filter employees based on these thresholds
filtered_df = merged_df[(merged_df['Salary'] > salary_threshold) & (merged_df['ExperienceYears'] > experience_threshold)]

filtered_df
```
