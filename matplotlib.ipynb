```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# simulated data; years and corresponding sales
a = np.arange(2010, 2021) # years 2010 tp 2021
b = np.array([100, 120, 130, 160, 180, 200, 240, 260, 280, 300, 320])

```

```python
# simulated data: age vs. blood pressure
x=np.arange(20, 70, 5)  # ages from 20 to 65
y = 0.8 * x + 80 #Blood Pressure vlaues
```

```python
# --- plot 1: Age vs. Blood Pressure ---

plt.figure(figsize =(8,5))
plt.plot(x,y,color = 'green', marker ='o', linestyle ='-', label ='BP vs Age')
plt.title("BP by Age")
plt.xlabel("Age in yrs")
plt.ylabel("Blood pressure in mmHg")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(a,b,color = 'blue', marker ='s', linestyle ='--', label ='Annual Sales')
plt.title("Annual Sales trend")
plt.xlabel("year")
plt.ylabel("Sales in 1000$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(14,5))

#plot 1: Blood Pressure vs. Age
axes[0].plot(x,y, color='green', marker='o', linestyle='-', label='BP vs Age')
axes[0].set_title('Blood Pressure by Age')
axes[0].set_xlabel('Age(years)')
axes[0].set_ylabel('Blood Pressure (mm Hg)')
axes[0].grid(True)
axes[0].legend()


# plot2: SLAES VS YEAR
axes[1].plot(a,b, color='blue', marker='s', linestyle='--', label='Annual Sales')
axes[1].set_title('Annual sales Trend')
axes[1].set_xlabel('Year')
axes[1].set_ylabel("Sales(in $1000s)")
axes[1].grid(True)
axes[1].legend()
plt.tight_layout()
plt.show()
```
