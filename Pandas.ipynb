```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

departments = ["Data Science", "HR", "Sales", "IT Support", "Marketing", "Finance", "Operations"]
locations = ["Nagpur", "Mumbai", "Delhi", "Bangalore", "Pune", "Hyderabad", "Chennai", "Kolkata"]

data = []
for emp_id in range(1001, 2001):
    name = fake.first_name()
    department = random.choice(departments)
    age = random.randint(22, 50)
    salary = random.randint(30000, 150000)
    experience = random.randint(0, min(age - 21, 25))
    location = random.choice(locations)
    data.append([emp_id, name, department, age, salary, experience, location])

df = pd.DataFrame(data, columns=["EmployeeID", "Name", "Department", "Age", "Salary", "ExperienceYears", "Location"])

# Add columns
df['Status'] = df['Salary'].apply(lambda x: 'High' if x > 80000 else 'Normal')
df['EligibleForPromotion'] = df['ExperienceYears'].apply(lambda x: 'Yes' if x >= 5 else 'No')
df['AgeGroup'] = df['Age'].apply(lambda x: 'Young' if x < 30 else ('Mid' if (x >= 30 and x < 40) else 'Senior'))

# Save CSVs to your working directory
df.to_csv("updated_employee_data.csv", index=False)
df[df['Status'] == 'High'].to_csv("high_salary_employees.csv", index=False)
df[df['AgeGroup'] == 'Young'].to_csv("young_employees.csv", index=False)

print("✅ All files saved successfully.")
```

```python
# Create a new DataFrame with performance scores
import pandas as pd
import random

# Assume 1000 employee IDs from 1001 to 2000
performance_data = {
    'EmployeeID': list(range(1001, 2001)),
    'PerformanceRating': [random.choice(['Excellent', 'Good', 'Average', 'Poor']) for _ in range(1000)],
    'LastPromotionYear': [random.choice([2023, 2022, 2021, 2020, None]) for _ in range(1000)]
}

performance_df = pd.DataFrame(performance_data)

# Save it to CSV
performance_df.to_csv("performance_data.csv", index=False)
```

```python
# prompt: 💼 Task 2: Rank Employees by Salary
# Add a new column called "SalaryRank" which gives:
# Rank 1 to the highest salary
# Ties should be handled by giving the same rank (method='min'), explain use of method min

# Add 'SalaryRank' column using rank() method
# method='min' assigns the same rank to rows with equal values (ties), which is the minimum of the ranks that would have been assigned without considering ties.
df['SalaryRank'] = df['Salary'].rank(method='min', ascending=False)

# Convert rank to integer for cleaner look
df['SalaryRank'] = df['SalaryRank'].astype(int)

print(df[['Name', 'Salary', 'SalaryRank']].sort_values(by='SalaryRank').head())
```

```python
# prompt: 💼 Task 4: Multi-column Grouping
# For each Department + AgeGroup, calculate:
# Count of employees
# Average Salary
# Show the top 10 rows only.( explain every process

# 1. Group by 'Department' and 'AgeGroup': We use the `groupby()` method on the DataFrame `df`.
# We provide a list `['Department', 'AgeGroup']` to group by multiple columns simultaneously.
# 2. Calculate Aggregations: After grouping, we apply the `agg()` method. This allows us to perform multiple aggregation functions in a single step.
# We provide a dictionary where keys are the columns we want to aggregate and values are the aggregation functions we want to apply.
# 'EmployeeID': 'count' calculates the number of employees in each group.
# 'Salary': 'mean' calculates the average salary for each group.
# We reset the index using `.reset_index()` to convert the grouped output (which has a multi-level index based on Department and AgeGroup) back into a standard DataFrame.
department_age_group_summary = df.groupby(['Department', 'AgeGroup']).agg(
    EmployeeCount=('EmployeeID', 'count'),  # Count of employees
    AverageSalary=('Salary', 'mean')        # Average Salary
).reset_index()

# 3. Sort the results (optional but often useful): We sort the resulting DataFrame to see groups with higher employee counts or average salaries first.
# We sort by 'EmployeeCount' in descending order (`ascending=False`).
department_age_group_summary = department_age_group_summary.sort_values(by='EmployeeCount', ascending=False)

# 4. Show the top 10 rows: We use the `.head(10)` method to display only the first 10 rows of the sorted DataFrame.
print("Summary by Department and Age Group (Top 10):")
print(department_age_group_summary.head(10))
```
