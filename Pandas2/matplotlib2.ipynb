```python
# pandas visulaization vs Matplolib vs Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart.csv')
df
```

```python
ages=df.groupby('age').median().reset_index()
ages
```

```python
# Simple Pandas
ages.plot.line('age', 'chol')
plt.show()
```

```python
# matplotlib
fig, ax = plt.subplots()
ax.plot(ages['age'], ages['chol'], ls=':', lw=1.7)  # line style , line width
ax.set_xlabel('Age')
ax.set_ylabel('Cholestrol')
plt.show()
```

```python
# seaborn

sns.lineplot(x='age', y='chol', data=ages, linestyle=':', linewidth=1.7)
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.show()
```
