```python
import pandas as pd
```

```python
df=pd.read_csv('titanic-tested.csv')
df
```

```python
print(df['Age'].describe())
```

```python
#count the number of null values in the age column
null_count = df['Age'].isnull().sum()
print(f"Number of null values in the age column:{null_count}")
```

```python
#1. Handle missing vlaues in the age column
# Replace missing values with the median of the age column
df['Age']=df['Age'].fillna(df['Age'].median())
df['Age']
```

```python
#2. Remove outliers in the age column (assuming ages>100 are outliers)
df = df[df['Age']<=100]
df
```

```python
# 3. Ensure the age column has consistent data types
df['Age']=df['Age'].astype(float)
df['Age']
```

```python
# Display cleaned info

print(df.info())
print(df['Age'].describe())
```

```python
# save the cleaned dataset (optional)
df.to_csv('titanic-tested_cleaned.csv', index=False)
```
